```python
import requests
import json

from pprint import pprint
```

```python
url = 'https://api.github.com'
user = 'erik-mv' 
req = requests.get(f'{url}/users/{user}/repos') 
r = req.json()
```

```python

print(r)
```

```python
repos = []
for i in r:
    repos.append(i['name'])
```

```python
print(repos)
```

```python
with open('Repos.json', 'w') as f:
    json.dump(repos, f)
```
